```python
import scipy.stats as stats
#Descriptive statistics of both post-molt of field
# Filtering the DataFrame for post_molt sizes of 'Field' in 'Location'
filtered_df_field = df[df['Location'] == 'Field']

# Calculate Maximum and Minimum values 
max_value = filtered_df_field['Post-molt'].max()
min_value = filtered_df_field['Post-molt'].min()

# Calculate Median, Mean, Standard deviation, Skewness, and Kurtosis
median_value = filtered_df_field['Post-molt'].median()
mean_value = filtered_df_field['Post-molt'].mean()
std_deviation = filtered_df_field['Post-molt'].std()
skewness = stats.skew(filtered_df_field['Post-molt'])
kurtosis = stats.kurtosis(filtered_df_field['Post-molt'])

# Print the results
print(f'Maximum Value: {max_value}')
print(f'Minimum Value: {min_value}')
print(f'Median: {median_value}')
print(f'Mean: {mean_value}')
print(f'Standard Deviation: {std_deviation}')
print(f'Skewness: {skewness}')
print(f'Kurtosis: {kurtosis}')
```

```python
from scipy.stats import linregress
# Perform linear least squares of field
slope, intercept, r_value, p_value, std_err = linregress(post_molt_sizes_field, pre_molt_sizes_field)

# Calculate predicted values using the linear model
predicted_pre_molt_sizes_field = slope * post_molt_sizes_field + intercept

# Calculate residuals
residuals_field = pre_molt_sizes_field - predicted_pre_molt_sizes_field

# Plot the data and regression line
plt.figure(figsize=(12, 6))

# Scatterplot with regression line
plt.subplot(1, 2, 1)
plt.scatter(post_molt_sizes_field, pre_molt_sizes_field, color='blue', alpha=0.7, label='Data')
plt.plot(post_molt_sizes_field, predicted_pre_molt_sizes_field, color='red', label='Linear Regression')

# Add labels and title
plt.title('Linear Least Squares: Pre-molt Sizes_field vs Post-molt Sizes_field')
plt.xlabel('Post-molt Sizes_field')
plt.ylabel('Pre-molt Sizes_field')
plt.legend()

# Residual plot
plt.subplot(1, 2, 2)
plt.scatter(predicted_pre_molt_sizes_field, residuals_field, color='green', alpha=0.7)
plt.axhline(y=0, color='black', linestyle='--', label='Zero Residuals Line')

# Add labels and title
plt.title('Residual of field Plot')
plt.xlabel('Predicted Pre-molt Sizes_field')
plt.ylabel('Residuals_field')
plt.legend()

# Show the plots
plt.tight_layout()
plt.show()

# Print Pearson's R-squared value
r_squared = r_value**2
print(f"Pearson's R-squared value: {r_squared}")
```

```python
import numpy as np
from sklearn.model_selection import cross_val_predict, KFold
from sklearn.metrics import r2_score
# Utilizing cross-validation to estimate prediction accuracy of linear models
# Combine the data for lab and field crabs
all_post_molt_sizes = pd.concat([post_molt_sizes_field, post_molt_sizes_lab], ignore_index=True)
all_pre_molt_sizes = pd.concat([pre_molt_sizes_field, pre_molt_sizes_lab], ignore_index=True)
all_locations = ['Field'] * len(post_molt_sizes_field) + ['Lab'] * len(post_molt_sizes_lab)

# Create a DataFrame for combined data
combined_data = pd.DataFrame({'Location': all_locations, 'Post-molt': all_post_molt_sizes, 'Pre-molt': all_pre_molt_sizes})

# Perform k-fold cross-validation for both lab and field crabs
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

# Initialize arrays to store predicted values and true labels
predicted_values = np.array([])
true_labels = np.array([])

# Loop through each fold
for train_idx, test_idx in kfold.split(combined_data):
    # Split data into training and testing sets for each fold
    train_data = combined_data.iloc[train_idx]
    test_data = combined_data.iloc[test_idx]

    # Fit the model on training data (use sm.OLS here)
    model = sm.OLS(train_data['Pre-molt'], sm.add_constant(train_data['Post-molt']))
    results = model.fit()

    # Predict on the testing data
    predictions = results.predict(sm.add_constant(test_data['Post-molt']))

    # Store the predicted values and true labels
    predicted_values = np.concatenate((predicted_values, predictions))
    true_labels = np.concatenate((true_labels, test_data['Pre-molt'].values))

# Calculate R-squared manually
r_squared = r2_score(true_labels, predicted_values)
print("Cross-validated R-squared:", r_squared)
```
